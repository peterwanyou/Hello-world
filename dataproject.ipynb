Introduction à Pandas



Pandas est une bibliothèque Python open source qui fournit des outils de manipulation et d'analyse de données hautes performances.
Le nom Pandas est dérivé du mot Panel Data, un terme économétrique pour les données multidimensionnelles.
Grâce à Pandas, nous pouvons accomplir les cinq étapes typiques du traitement et de l'analyse des données, quelle que soit l'origine des données : charger, préparer, manipuler, modéliser et analyser.
Principales fonctionnalités de Pandas
Objet DataFrame rapide et efficace avec indexation par défaut et personnalisée.
Outils de chargement de données dans des objets de données en mémoire à partir de différents formats de fichiers.
Alignement des données et gestion intégrée des données manquantes.
Remodelage et pivotement des ensembles de données.
Découpage, indexation et sous-ensemble basés sur des étiquettes de grands ensembles de données.
Les colonnes d'une structure de données peuvent être supprimées ou insérées.
Regrouper par données pour l'agrégation et les transformations.
Fusion et jointure de données hautes performances.
Fonctionnalité de séries temporelles.
Structures de données dans Pandas
Série : unidimensionnelle
Trame de données : bidimensionnelle
Panneau : tridimensionnelle et plus

```python
# extraire tous les enregistrements où la marque est « Close-up »

df = pos_data.loc[(pos_data['Brand']=='Close-up')]
df.head()
```

```python
# Afficher le nombre de ligne que nous avons dans df
df.shape[0]
```

```python
# extraire tous les enregistrements dont les revenus sont compris entre 10 000 et 15 000

df_1 = pos_data.loc[(pos_data['Revenue($)']>=10000) & (pos_data['Revenue($)']<=15000)]
pos_data.head()
```

```python
# Affichons le nombre d'element dans cette condition df_1

df_1.shape[0]
```

```python
# Extraire les revenue superieur a 25000 et les traffic sur les page inferieur a 1000

pos_data = pos_data.loc[(pos_data['Revenue($)']>=25000) & (pos_data['Page_traffic']<1000)]
pos_data.head()
```

```python
# verifions les different années sur lesquelles sont enregistré nos données

pos_data['Date'].dt.year.unique().tolist()
```

```python
# Nous allons regrouper et calaculer la moyen les Revenue($),	Units_sold,	Page_traffic en fonction de l'année

df = pos_data.groupby(pos_data['Date'].dt.year).mean(numeric_only=True).round(2)
df.head()
```

```python
#créer un sous-ensemble de données où l'année est 2021 et le secteur est Oral Care.

df = pos_data.loc[(pos_data['Date'].dt.year==2021) & (pos_data['Sector']=='Oral Care')]
df
```

```python
# definir le nombre de ligne

df.shape[0]
```

```python
#maintenant, regroupons le sous-ensemble en fonction du mois et découvrons le revenu moyen

df = df.groupby(df['Date'].dt.month).mean(numeric_only=True).round(2).drop(['Units_sold','Page_traffic'], axis=1)
df.head()
```

```python
# Remplacons la date par mois et recvenue par AVG revenue

df = df.reset_index().rename(columns={'Date':'Month', 'Revenue($)': 'AVG revenue'})
df.head()
```

```python
# Definir la colonne Month comme notre index

df = df.set_index('Month')
df.head()
```

```python
# importons la librairie calendrier

import calendar
```

```python
# ecrire une fonction pour definir le nom du mois

def monthname(x):
  return calendar.month_name[x]
```

```python
# Reset the index so 'Month' becomes a column again
df = df.reset_index()
# Remplacons lemois numeric par le nom reel du mois
df['Month'] = df['Month'].apply(monthname)
# Set 'Month' as the index again if desired
df = df.set_index('Month')
df.head()
```
